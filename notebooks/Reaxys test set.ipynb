```python
import pandas as pd
```

```python
df = pd.read_csv('../data/DA_regio_solvent.csv')
df
```

```python
with open('../data/DA_test_no_solvent.txt', 'w') as f:
    f.writelines("\n".join(df['reaction_smiles'].values))       
```

```python
lines = []
for reaction, xtb_s, orca_s in zip(
    df["reaction_smiles"].values,
    df["xtb_solvent"].values,
    df["orca_solvent"].values
):
    lines.append(f"{reaction} {xtb_s} {orca_s}")

with open('../data/DA_test_solvent.txt', 'w') as f:
    f.writelines("\n".join(lines))   
```
