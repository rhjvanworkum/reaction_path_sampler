```python
import pandas as pd
import os
```

```python
df = pd.read_csv('../data/DA_regio_solvent.csv')
df
```

### Generate txt files with reaction smiles

```python
with open('../data/DA_test_solvent.txt', 'w') as f:
    f.writelines("\n".join(df['reaction_smiles'].values))       
```

```python
lines = []
for reaction, xtb_s, orca_s in zip(
    df["reaction_smiles"].values,
    df["xtb_solvent"].values,
    df["orca_solvent"].values
):
    lines.append(f"{reaction} {xtb_s} {orca_s}")

with open('../data/DA_test_solvent.txt', 'w') as f:
    f.writelines("\n".join(lines))   
```

### create new DataFrame containing only succesfully simulated reactions

```python
success_uids = []
for root, dirs, files in os.walk('../scratch/DA_test_solvent/'):
    if len(root.split('/')) > 3 and root.split('/')[-2] == 'DA_test_solvent':
        if os.path.exists(os.path.join(root, 'reaction.xyz')):
            success_uids.append(int(root.split('/')[-1]))
success_uids = sorted(success_uids)
print(success_uids)
```

```python
uids_to_keep = []
for reaction_idx in df['reaction_idx'].unique():
    selection = df[df['reaction_idx'] == reaction_idx]
    if len(selection[selection['uid'].isin(success_uids)]) > 1:
        for uid in selection[selection['uid'].isin(success_uids)]['uid']:
            uids_to_keep.append(uid)
            
df = df[df['uid'].isin(uids_to_keep)]
df
```

```python
df.to_csv('../data/DA_regio_solvent_success.csv')
```
