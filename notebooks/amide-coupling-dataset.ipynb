```python
import pandas as pd

from rdkit import Chem
```

```python
df = pd.read_csv('../data/ac_dataset_dcm_small.csv')
df
```

```python
df['reaction_smiles'] = [
    f"{Chem.MolToSmiles(Chem.MolFromSmiles(subs))}>>{Chem.MolToSmiles(Chem.MolFromSmiles(prods))}"
    for subs, prods in zip(df['substrates'], df['products'])
]
df = df.drop_duplicates(subset="reaction_smiles")
df
```

```python
df.to_csv('../data/ac_dataset_dcm_small.csv')
```
